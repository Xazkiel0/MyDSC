```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="darkgrid")
```

```python
df = pd.read_excel("K!.xlsx",parse_dates=["TANGGAL LAHIR"],date_parser=pd.to_datetime)
df.head()
```

```python
df.isnull().sum()
df.dropna(inplace=True)
df["TEMPAT LAHIR"] = df["TEMPAT LAHIR"].apply(lambda x: str(x).upper())
df["ASAL SEKOLAH"] = df["ASAL SEKOLAH"].apply(lambda x: str(x).upper())
```

```python
tkj = df[df.JURUSAN == "TKJ"]
rpl = df[df.JURUSAN == "RPL"]
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data=tkj,y="ASAL SEKOLAH",order=tkj["ASAL SEKOLAH"].value_counts().index)
plt.show()
```

```python
plt.figure(figsize=(12,10))
sns.countplot(data=rpl,y="ASAL SEKOLAH",order=rpl["ASAL SEKOLAH"].value_counts().index)
plt.show()
```
