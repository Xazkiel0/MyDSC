```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_excel("Data - Hasil Survey (1).xlsx");

df.drop(columns=["Unnamed: 15"],inplace=True)

df.info()
```

```python
df.head(10)
```

```python
def yorn(word):
    return 1 if word == "Ya" else 0

df["Kuliah"] = df["Kuliah"].apply(yorn,int)
df["Bisnis Mandiri"] = df["Bisnis Mandiri"].apply(yorn,int)
df["Bekerja"] = df["Bekerja"].apply(yorn,int)
df["Tempat Kerja"].fillna("Tidak Punya",inplace=True)
df["Tempat Kuliah"].fillna("Tidak Punya",inplace=True)
df["Bidang Kuliah"].fillna("Tidak Ada",inplace=True)
df["Bisnis"].fillna("Tidak Punya",inplace=True)
df["Job"].fillna("Bukan Pekerja",inplace=True)
df.head()
```

```python
def filterAsal(asal):
    return "SMKN 1 Boyolangu" if asal[:10].lower() in ["smkn 1 boy","smkn 2 boy","smkn 3 boy"]  else asal

df["Asal SMK/SMA"] = df["Asal SMK/SMA"].apply(filterAsal)
```

```python
# def singkatan(words):
#     shard = words.split(' ')
#     return ''.join(w[0] if len(shard) > 1 else w for w in shard )

df["Tempat Kuliah"] = df["Tempat Kuliah"].apply(lambda x: x.upper())
df["Jurusan"] = df["Jurusan"].apply(lambda x: x.upper())
df["Asal SMP"] = df["Asal SMP"].apply(lambda x: x.upper())
df["Passion"] = df["Passion"].apply(lambda x: x.capitalize())
df["Ekskul"] = df["Ekskul"].apply(lambda x: x.capitalize())
```

```python
plt.figure(figsize=(16,6))
ax = sns.countplot(y="Jurusan",hue="Kuliah", data=df[["Jurusan","Kuliah"]], order=df.Jurusan.value_counts().index)
```

```python
plt.figure(figsize=(16,6))
ax = sns.countplot(y="Jurusan",hue="Bekerja", data=df,order=df.Jurusan.value_counts().index)
```

```python
def pieVal(column):
    labels = df[column].unique()[df[column].value_counts() > 5].tolist()
    labels.append("Lainnya")
    values = df[column].value_counts()[df[column].value_counts() > 5].tolist()
    others = df[column].value_counts()[df[column].value_counts() <= 5].sum()
    values.append(others)
    return labels,values

labels,values = pieVal("Bidang Kuliah")
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```

```python
labels,values = pieVal("Jurusan")
fig1, ax1 = plt.subplots()
ax1.pie(values, labels=labels, autopct='%1.1f%%', shadow=True)
plt.show()
```
